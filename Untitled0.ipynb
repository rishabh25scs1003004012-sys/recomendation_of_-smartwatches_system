```python
import pandas as pd
```

```python
df = pd.read_csv('student_performance.csv')
```

```python
# Kitne rows aur columns hain?
print(df.shape)

# Pehli 5 rows dekho
print(df.head())

# Har column ka data type
print(df.dtypes)
```

```python
# Kitne duplicates hain?
print("Duplicates:", df.duplicated().sum())

# Delete karo
df = df.drop_duplicates()
print("Duplicates removed! Rows ab:", len(df))
```

```python
# Konse columns bilkul empty hain?
blank_cols = [c for c in df.columns if df[c].isna().all()]
print("Blank columns:", blank_cols)

# Delete karo
df = df.drop(columns=blank_cols)
```

```python
# Har column mein kitni missing values hain?
print(df.isnull().sum())

# Option A: Missing rows delete karo
df = df.dropna()

# Option B: Missing values fill karo (0 se ya average se)
df = df.fillna(0)           # 0 se fill karo
df = df.fillna(df.mean())   # average se fill karo
```

```python
df = df.reset_index(drop=True)
```

```python
df.to_csv('cleaned_data.csv', index=False)
print("File save ho gayi! ✅")
```

```python
import pandas as pd

# 1. Load
df = pd.read_csv('student_performance.csv')
print("Before cleaning:", df.shape)

# 2. Duplicates remove
df = df.drop_duplicates()

# 3. Blank columns remove
blank_cols = [c for c in df.columns if df[c].isna().all()]
df = df.drop(columns=blank_cols)

# 4. Missing values handle
df = df.dropna()

# 5. Index reset
df = df.reset_index(drop=True)

print("After cleaning:", df.shape)

# 6. Save
df.to_csv('cleaned_data.csv', index=False)
print("Done! ✅")
```
